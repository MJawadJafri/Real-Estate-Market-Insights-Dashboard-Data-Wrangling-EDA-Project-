```python
# prompt: perform eda

# Imports for plotting
import matplotlib.pyplot as plt
import seaborn as sns

# Display basic information about the dataframe
print("Data Info:")
data.info()

# Display descriptive statistics
print("\nDescriptive Statistics:")
print(data.describe())

# Function to convert price to millions
def convert_price_to_millions(price_str):
    price_str = price_str.replace(',', '') # Remove commas
    if 'Crore' in price_str:
        return float(price_str.replace('Crore', '').strip()) * 10
    elif 'Lakh' in price_str:
        return float(price_str.replace('Lakh', '').strip()) * 0.1
    else:
        try:
            return float(price_str.strip()) / 1000000 # Convert to millions for smaller values
        except ValueError:
            return None # Handle other values that cannot be converted

# Apply the conversion function to the 'price' column
data['price_in_millions'] = data['price'].apply(convert_price_to_millions)


# Visualize the distribution of the 'price' column
plt.figure(figsize=(10, 6))
sns.histplot(data['price_in_millions'].dropna(), kde=True) # Use the new numeric column
plt.title('Distribution of Property Prices (in Millions PKR)')
plt.xlabel('Price (Millions PKR)')
plt.ylabel('Frequency')
plt.show()



# Boxplot of prices by city
plt.figure(figsize=(12, 8))
sns.boxplot(x='location_city', y='price_in_millions', data=data.dropna(subset=['price_in_millions'])) # Use the new numeric column
plt.title('Price Distribution by City (in Millions PKR)')
plt.xlabel('City')
plt.ylabel('Price (Millions PKR)')
plt.xticks(rotation=45, ha='right')
plt.show()

# Count plot of property types
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=data)
plt.title('Count of Property Types')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.show()

# Count plot of purposes
plt.figure(figsize=(8, 6))
sns.countplot(x='purpose', data=data)
plt.title('Count of Property Purposes')
plt.xlabel('Purpose')
plt.ylabel('Count')
plt.show()
```

```python
# prompt: find correation between values

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Function to convert area to square feet
def convert_area_to_sqft(area_str):
    try:
        area_str = str(area_str).strip().lower()
        if 'kanal' in area_str:
            value = float(area_str.replace('kanal', '').strip())
            return value * 5445.0  # 1 Kanal = 5445 Sqft
        elif 'marla' in area_str:
            value = float(area_str.replace('marla', '').strip())
            return value * 272.25  # 1 Marla = 272.25 Sqft
        elif 'sqft' in area_str:
            return float(area_str.replace('sqft', '').strip())
        elif 'sqyd' in area_str:
            value = float(area_str.replace('sqyd', '').strip())
            return value * 9.0 # 1 SqYd = 9 Sqft
        else:
            return None  # Handle other values that cannot be converted
    except ValueError:
        return None # Handle potential errors during conversion

# Apply the conversion function to the 'area' column
data['area_in_sqft'] = data['area'].apply(convert_area_to_sqft)

# Convert 'bedroom' column to numeric, coercing errors to NaN
data['bedroom'] = pd.to_numeric(data['bedroom'], errors='coerce')


# Select relevant numerical columns for correlation analysis, skipping 'area_in_sqft'
numerical_cols = ['bedroom', 'price_in_millions']

# Drop rows with NaN values in the selected numerical columns before calculating correlation
data_for_corr = data[numerical_cols].dropna()

# Calculate the correlation matrix
correlation_matrix = data_for_corr.corr()

# Display the correlation matrix
print("\nCorrelation Matrix:")
display(correlation_matrix)

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features (excluding area)')
plt.show()
```

```python
# prompt: find outlier

# Finding Outliers using Z-score method on 'price_in_millions'

# Calculate Z-scores for the 'price_in_millions' column
# We use a copy to avoid SettingWithCopyWarning if we were to modify the original column later
data_with_zscore = data.copy()
data_with_zscore['price_zscore'] = (data_with_zscore['price_in_millions'] - data_with_zscore['price_in_millions'].mean()) / data_with_zscore['price_in_millions'].std()

# Define a threshold for outlier detection (e.g., Z-score > 3 or < -3)
zscore_threshold = 3

# Identify rows where the Z-score of 'price_in_millions' exceeds the threshold
outliers_zscore = data_with_zscore[(data_with_zscore['price_zscore'].abs() > zscore_threshold)]

print(f"\nOutliers detected using Z-score (threshold = {zscore_threshold}):")
display(outliers_zscore[['price', 'price_in_millions', 'price_zscore']])

# Finding Outliers using IQR method on 'price_in_millions'

# Calculate the first quartile (Q1) and third quartile (Q3)
Q1 = data['price_in_millions'].quantile(0.25)
Q3 = data['price_in_millions'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outlier detection
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify rows where 'price_in_millions' falls outside the IQR bounds
outliers_iqr = data[(data['price_in_millions'] < lower_bound) | (data['price_in_millions'] > upper_bound)]

print(f"\nOutliers detected using IQR:")
display(outliers_iqr[['price', 'price_in_millions']])

# Note: Depending on the dataset and the context, you might choose one method over the other,
# or use visualization (like the boxplot already shown) to visually identify potential outliers.
# The Z-score method assumes a normal distribution, while IQR is less sensitive to the distribution shape.
```
